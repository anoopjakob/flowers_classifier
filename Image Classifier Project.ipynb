````python
# RUN THIS AFTER LOADING THE CHECKPOINT

# ERROR CLEARED criterion used as NLLL since we are using log_softmax for output
criterion = nn.NLLLoss()

#for p in classifier.parameters():
#    p.requires_grad = True

#while using optimizer i am getting error ' optimizing a parameter that doesnt require gradinets'
# so i find this work around . dont know if its going to work

#i understood the abov problem.. my order of freezing and creating new classifier was wrong 

# usig adam optimizer
optimizer = optim.Adam(model.classifier.parameters(), lr = 0.0009)
# trying this learining rate.. if not effective .will change

````

````python
# ALREADY TRAINED .. DONT RUN THIS CODE AGAIN

# OR RUN THIS AFTER LOADING THE CHECKPOINT

# implementing validation as a function to reduce the crowd
# so this function do the validation pass

def validation(model, validloader, criterion):
    valid_loss = 0
    accuracy = 0
    
    # change model to work with cuda
    model.to('cuda')

    # iteration ..here getting data using valid loader
    for ii, (images, labels) in enumerate(validloader):
    
        # changing below to work with cuda environment for making it faster
        # edited this code in commandline application since this will return an
        # error if program not running in gpu
        images, labels = images.to('cuda'), labels.to('cuda')

        # Forward pass image though model for prediction
        output = model.forward(images)
        
        # Calculating loss
        valid_loss += criterion(output, labels).item()
        
        # Calculate probability
        ps = torch.exp(output)
        
        # CalculatIN THE accuracy
        equality = (labels.data == ps.max(dim=1)[1])
        accuracy += equality.type(torch.FloatTensor).mean()
    
    return valid_loss, accuracy
````

````python
# ALREADY TRAINED .. DONT RUN THIS CODE..OR FIRST LOAD THE CHECKPOINT AND THEN RUN 

# THE TRAINING AREA

epochs = 3
steps = 0
print_every = 40

# change to gpu mode
model.to('cuda')

for e in range(epochs):
    # using time module to know how much time will it take for 3 epochs
    start = time.time()
    running_loss = 0
    
    # Iterating over data to carry out training step
    for ii, (inputs, labels) in enumerate(trainloader):
        steps += 1
        
        inputs, labels = inputs.to('cuda'), labels.to('cuda')
        
        # zeroing parameter gradients
        optimizer.zero_grad()
        
        # Forward and backward passes
        outputs = model.forward(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        
        # Carrying out validation step
        #validation does not takes place in every loop..
        # only in between
        if steps % print_every == 0:
            # setting model to evaluation mode
            model.eval()
            
            # Gradients are turned OFF  we are evaluating
            with torch.no_grad():
                valid_loss, accuracy = validation(model, validloader, criterion)
                
            print(f"No. epochs: {e+1}, \
            Training Loss: {round(running_loss/print_every,3)} \
            Valid Loss: {round(valid_loss/len(validloader),3)} \
            Valid Accuracy: {round(float(accuracy/len(validloader)),3)}")
            
            running_loss = 0
            
            # Turning training back on
            model.train()
            
    total = time.time() - start
    print('time taken {}:{}:{}'.format( int(total/3600), int((total%3600)/60),int((total%3600)%60) ))
            
            
````

## Image Preprocessing

You'll want to use `PIL` to load the image ([documentation](https://pillow.readthedocs.io/en/latest/reference/Image.html)). It's best to write a function that preprocesses the image so it can be used as input for the model. This function should process the images in the same manner used for training. 

First, resize the images where the shortest side is 256 pixels, keeping the aspect ratio. This can be done with the [`thumbnail`](http://pillow.readthedocs.io/en/3.1.x/reference/Image.html#PIL.Image.Image.thumbnail) or [`resize`](http://pillow.readthedocs.io/en/3.1.x/reference/Image.html#PIL.Image.Image.thumbnail) methods. Then you'll need to crop out the center 224x224 portion of the image.

Color channels of images are typically encoded as integers 0-255, but the model expected floats 0-1. You'll need to convert the values. It's easiest with a Numpy array, which you can get from a PIL image like so `np_image = np.array(pil_image)`.

As before, the network expects the images to be normalized in a specific way. For the means, it's `[0.485, 0.456, 0.406]` and for the standard deviations `[0.229, 0.224, 0.225]`. You'll want to subtract the means from each color channel, then divide by the standard deviation. 

And finally, PyTorch expects the color channel to be the first dimension but it's the third dimension in the PIL image and Numpy array. You can reorder dimensions using [`ndarray.transpose`](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.ndarray.transpose.html). The color channel needs to be first and retain the order of the other two dimensions.

````python
def process_image(image):
    ''' Scales, crops, and normalizes a PIL image for a PyTorch model,
        returns an Numpy array
    '''
    
    # DONE: Process a PIL image for use in a PyTorch model

    # Converting image to PIL image
    pil_im = Image.open(f'{image}' + '.jpg')

    # transforming
    transform = transforms.Compose([transforms.Resize(256),
                                    transforms.CenterCrop(224),
                                    transforms.ToTensor(),
                                    transforms.Normalize([0.485, 0.456, 0.406], 
                                                         [0.229, 0.224, 0.225])]) 
    
    pil_tfd = transform(pil_im)
    
    # Converting to Numpy array 
    array_im_tfd = np.array(pil_tfd)
    
    return array_im_tfd
   
````

To check your work, the function below converts a PyTorch tensor and displays it in the notebook. If your `process_image` function works, running the output through this function should return the original image (except for the cropped out portions).

````python
def imshow(image, ax=None, title=None):
    """Imshow for Tensor."""
    if ax is None:
        fig, ax = plt.subplots()
    
    # PyTorch tensors assume the color channel is the first dimension
    # but matplotlib assumes is the third dimension
    image = image.numpy().transpose((1, 2, 0))
    
    # Undo preprocessing
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    image = std * image + mean
    
    # Image needs to be clipped between 0 and 1 or it looks like noise when displayed
    image = np.clip(image, 0, 1)
    
    ax.imshow(image)
    
    return ax
````

## Class Prediction

Once you can get images in the correct format, it's time to write a function for making predictions with your model. A common practice is to predict the top 5 or so (usually called top-$K$) most probable classes. You'll want to calculate the class probabilities then find the $K$ largest values.

To get the top $K$ largest values in a tensor use [`x.topk(k)`](http://pytorch.org/docs/master/torch.html#torch.topk). This method returns both the highest `k` probabilities and the indices of those probabilities corresponding to the classes. You need to convert from these indices to the actual class labels using `class_to_idx` which hopefully you added to the model or from an `ImageFolder` you used to load the data ([see here](#Save-the-checkpoint)). Make sure to invert the dictionary so you get a mapping from index to class as well.

Again, this method should take a path to an image and a model checkpoint, then return the probabilities and classes.

```python
probs, classes = predict(image_path, model)
print(probs)
print(classes)
> [ 0.01558163  0.01541934  0.01452626  0.01443549  0.01407339]
> ['70', '3', '45', '62', '55']
```

````python
def predict(image_path, model, topk=5):
    ''' Predict the class (or classes) of an image using a trained deep learning model.
    '''
    
    # DONE: Implement the code to predict the class from an image file   
    
    # here since there is not training needed loading with cpu is enough
    loaded_model = load_checkpoint(model).cpu()
    
    # image should be converted to a certain type so that the model recognizes it
    # this is done using the process_image function created above
    img = process_image(image_path)
    
    # in process_image we changed the image values to numpy arrays but here to 
    # work with the rest of torch tensors we need to convert the numpy array to torch tensor
    img_to_tensor = torch.from_numpy(img).type(torch.FloatTensor)
    
    # Adding dimension to image to comply with (B x C x W x H) input of model
    
    img_adding_dim = img_to_tensor.unsqueeze_(0)

    # model is now in evaluation mode.. here we dont need to alter the gradients or train..
    loaded_model.eval()
    with torch.no_grad():
        # Running image through network
        output = loaded_model.forward(img_adding_dim)

    # Calculating probabilities
    # here since top_k variable is 5 by default unless we specify it will return top 5 probabilities
    probability = torch.exp(output)
    probs_top = probability.topk(topk)[0]
    index_top = probability.topk(topk)[1]
    print(index_top.type())
    
    # here we can see that the above values are inside the long tensor
    # we are changing it into a numpuy array to easily show in plots or lists
    # Converting probabilities and outputs to lists
    probs_top_list = np.array(probs_top)[0]
    # this is returned back
    
    index_top_list = np.array(index_top[0])
    
    # Loading index and class mapping as as shown in the above markdown
    class_to_idx = loaded_model.class_to_idx
    
    # Invert it to the dictionary so you get a mapping from index to class as well.
    indx_to_class = {x: y for y, x in class_to_idx.items()}

    # Converting index list to class list
    classes_top_list = []
    for index in index_top_list:
        classes_top_list += [indx_to_class[index]]
        
    return probs_top_list, classes_top_list
````
